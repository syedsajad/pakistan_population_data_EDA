```python
rural_PP=df['ALL SEXES (RURAL)'].sum() # total rural population of pakistan in 2017
print("The Total rural population of pakistan in 2017 was:", rural_PP)
```

```python
# How much Rural population increase in pakistan from 1998 - 2017?
rural_pp_1998=df['POPULATION 1998 (RURAL)'].sum()
rural_pp_2017=df['ALL SEXES (RURAL)'].sum()
print("The rural population increase in pakistan from 1998 to 2017:",rural_pp_2017-rural_pp_1998)
```

```python
# percentage of increased population
per_of_pp=rural_pp_2017/rural_pp_1998*100
print("Percentage of increased population is: ",per_of_pp)
```

```python
# How much Urban population increase in pakistan from 1998 - 2017?
Urban_pp_1998=df['POPULATION 1998 (URBAN)'].sum()
Urban_pp_2017=df['ALL SEXES (URBAN)'].sum()
print("The Urban population increase in pakistan from 1998 to 2017:",Urban_pp_2017-Urban_pp_1998)
```
